````python
import pandas as pd
import requests
import pymongo
import time
from bs4 import BeautifulSoup as soup
from splinter import Browser
from pprint import pprint
from webdriver_manager.chrome import ChromeDriverManager
import requests
````

````python
# Setup splinter and open the website

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

url = 'https://redplanetscience.com/'
browser.visit(url)
````

````python
# use BeatifulSoup to parse the HTML content 
soup = soup(browser.html, 'html.parser')
````

````python
# print the html parsed data
print(soup.prettify())
````

````python
# find and save teh latest title in a variable

news_title = soup.find('div', class_="content_title").text
news_title

````

````python
news_p =soup.find('div', class_="article_teaser_body").text
news_p
````

````python
# close browser
browser.quit()
````

```python
# Example:

```

````python

from bs4 import BeautifulSoup as soup
````

````python

# Setup splinter 

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

url_img = 'https://spaceimages-mars.com/'
browser.visit(url_img)
    
````

````python
 # Scrape page into Soup
html = browser.html
soup = soup(html, "html.parser")
````

````python
relative_image_url = soup.find_all('img', class_='headerimage')


````

````python
for img in relative_image_url:
    if img.has_attr('src'):
        img1=img['src']
        print(img['src'])
````

````python
featured_image_url = url_img+img1
featured_image_url
````
